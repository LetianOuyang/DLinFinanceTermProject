```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('./cache/FactorAndReturn.csv', low_memory=False)
```

```python
df.dropna(subset=['permno','RET','yyyymm'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.drop_duplicates(subset=['permno', 'yyyymm'], inplace=True)
```

```python
# 
def make_balanced_panel(df):
    #  permno  yyyymm
    permnos = df['permno'].unique()
    yyyymms = df['yyyymm'].unique()

    #  DataFrame permno  yyyymm 
    balanced_df = pd.MultiIndex.from_product([permnos, yyyymms], names=['permno', 'yyyymm']).to_frame(index=False)

    # 
    balanced_df = pd.merge(balanced_df, df, on=['permno', 'yyyymm'], how='left')

    return balanced_df
```

```python
df_balance = make_balanced_panel(df)
```

```python
#  RET  1000 
# RET
missing_counts = df_balance.groupby('permno')['RET'].apply(lambda x: x.isna().sum())

# RET 1000 
top_1000_stocks = missing_counts.nsmallest(1000).index

# 1000
df_balance_filter = df_balance[df_balance['permno'].isin(top_1000_stocks)]
```

```python
from scipy.interpolate import PchipInterpolator
from tqdm.notebook import tqdm
import numpy as np
import pandas as pd

# PCHIP
# DivInit IndIPO ExchSwitch DivOmit Spinoff 
# PCHIP
backfill_columns = ['DivInit', 'IndIPO', 'ExchSwitch', 'DivOmit', 'Spinoff']

# 
spline_columns = [col for col in df_balance_filter.columns 
                  if col not in ['permno', 'yyyymm'] + backfill_columns]

# 
df_interpolated = df_balance_filter.copy()

def pchip_interpolate_with_fill(series):
    """
    PCHIP 
    """
    if series.isna().all():
        # 
        return series
    
    # 
    series_filled = series.fillna(method='ffill').fillna(method='bfill')
    
    # PCHIP
    if series_filled.isna().any():
        valid_mask = ~series_filled.isna()
        valid_indices = series_filled.index[valid_mask]
        valid_values = series_filled.values[valid_mask]
        
        if len(valid_values) >= 2:
            try:
                # PCHIP
                f = PchipInterpolator(valid_indices, valid_values, extrapolate=False)
                interpolated = f(series.index)
                
                # Series
                result = pd.Series(interpolated, index=series.index)
                
                # 
                result = result.fillna(method='ffill').fillna(method='bfill')
                
                return result
            except:
                # PCHIP
                return series_filled.interpolate(method='linear').fillna(method='ffill').fillna(method='bfill')
        else:
            # 
            return series_filled
    else:
        return series_filled

# 
skipped_stocks = set()

for permno in tqdm(df_balance_filter['permno'].unique(), colour='blue', desc='Processing stocks'):
    mask = df_balance_filter['permno'] == permno
    stock_data = df_balance_filter[mask].copy()
    
    # 
    skip_stock = False
    
    # 
    for col in backfill_columns:
        if col in stock_data.columns:
            df_interpolated.loc[mask, col] = stock_data[col].ffill().bfill()
    
    # PCHIP
    for col in spline_columns:
        if col not in stock_data.columns:
            continue
            
        series = stock_data[col]
        
        # 
        if series.isna().all():
            skip_stock = True
            break
        
        # PCHIP
        interpolated = pchip_interpolate_with_fill(series)
        df_interpolated.loc[mask, col] = interpolated.values
    
    if skip_stock:
        skipped_stocks.add(permno)

# 
if skipped_stocks:
    print(f" {len(skipped_stocks)} ")
    df_interpolated = df_interpolated[~df_interpolated['permno'].isin(skipped_stocks)]

print("")
```
